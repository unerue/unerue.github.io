## 넘파이

* ndim: 차원(축)의 수
* shape: 각 차원의 크기(튜플)
* size: 전체 요소의 개수, shape의 각 항목의 곱
* dtype: 요소의 데이터 타입
* itemsize: 각 요소의 바이트 크기

```python
import cv2

image = cv2.imread('./img/cat-01.jpg')
print(type(image))
print(image.ndim)
print(image.shape)
print(image.size)
print(image.dtype)
print(image.itemsize)
```

## Matplotlib

Matplotlib은 데이터 시각화 라이브러리입니다. 파이썬은 해당 라이브러리를 이용하여 도표나 차트 등을 그릴 수 있습니다. 이미지 프로세싱이나 컴퓨터 비전 분야에서 여러 이미지를 화면에 띄우고 싶을 때 `opencv`와 함께 사용합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 6, 9, 2, 10, 12, 4, 5, 2])
plt.plot(a)
plt.show()
```

```python
x = np.arange(10)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(10)
y = x**2
plt.plot(x, y, 'r')
plt.show()
```

```python
x = np.arange(20)
f1 = x * 5
f2 = x ** 2
f3 = x**2 + x*2

plt.plot(x, '--r')
plt.plot(f1, '.g')
plt.plot(f2, 'vb')
plt.plot(f3, 'sk')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)

plt.subplot(2, 2, 1)
plt.plot(x, x**2)

plt.subplot(2, 2, 2)
plt.plot(x, x*5)

plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x))
plt.show()
```
